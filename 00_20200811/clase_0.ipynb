#### Ejercicio 9

```python
lista_carreras = df.dropna(subset=['Carrera'])['Carrera'].unique()
prom_c = []

for carrera in lista_carreras:
    filtro_carrera = df["Carrera"] == carrera
    gente_carrera = df[filtro_carrera]
    prom_c.append([carrera, gente_carrera['Promedio'].mean()])
    #print("Promedio",carrera,":",gente_carrera['Promedio'].mean())

print(pd.DataFrame(prom_c,columns=["Carrera","Promedio"]).sort_values('Promedio'))
```

La carrera con promedio más alto es Biología, la carrera con promedio más bajo es Paleontología. Por supuesto que esto no alcanza para hacer afirmaciones sobre el esfuerzo de los estudiantes, sólo nos dice que en nuestro subset de datos (y no en general, porque estamos trabajando con un dataset no representativo (poca gente)) los estudiantes de ciencias biológicas tienden a tener mejores promedios que los de paleontología (pero puede haber casos donde el promedio de alguien de paleontología sea mayor que el de alguien de ciencias biológicas).
