### Punto 3

#### Ejercicio 12

```python
def predigo_promedio_movil(X,Y,x_nuevo,ventana):
    filtro_alturas = (X>=x_nuevo-ventana) & (X<=x_nuevo+ventana)
    return Y[filtro_alturas].mean()
```

```python
predigo_promedio_movil(df['altura_madre'],df['altura'],156.5,2)
```

#### Ejercicio 13

```python
sns.scatterplot(data=df,x='altura_madre',y='altura')
prom_movil = [predigo_promedio_movil(df['altura_madre'],df['altura'],x_nuevo,1) for x_nuevo in df['altura_madre']]
sns.lineplot(x=df['altura_madre'],y=prom_movil, color='orange')
plt.legend(["Promedios m√≥viles alturas madres con h=1"])
```

#### Ejercicio 14 y 15

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df,x='altura_madre',y='altura')
prom_movil = [predigo_promedio_movil(df['altura_madre'],df['altura'],x_nuevo,1) for x_nuevo in df['altura_madre']]
sns.lineplot(x=df['altura_madre'],y=prom_movil, color="orange")
prom_movil = [predigo_promedio_movil(df['altura_madre'],df['altura'],x_nuevo,2) for x_nuevo in df['altura_madre']]
sns.lineplot(x=df['altura_madre'],y=prom_movil, color="red", alpha=0.5)
prom_movil = [predigo_promedio_movil(df['altura_madre'],df['altura'],x_nuevo,10) for x_nuevo in df['altura_madre']]
sns.lineplot(x=df['altura_madre'],y=prom_movil, color='green')
plt.legend(["h=1","h=2","h=10"])
```

### Punto 4

#### Ejercicio 16

```python
# X es columna de variables predictores, Y es columna de sus correspondientes 
# variables respuesta, x_nuevo es lugar donde predecir y k la cantidad de vecinos
def predigo_promedio_vecinos(X,Y,x_nuevo,k):
    X_cercanos = (X-x_nuevo).abs().sort_values()[:k] #Me da los INDICES de los mas cercanos!
    return Y[X_cercanos.index].mean()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df,x='altura_madre',y='altura')
prom_vecinos = [predigo_promedio_vecinos(df['altura_madre'],df['altura'],x_nuevo,4) for x_nuevo in df['altura_madre']]
sns.lineplot(x=df['altura_madre'],y=prom_vecinos, color="orange")
prom_vecinos = [predigo_promedio_vecinos(df['altura_madre'],df['altura'],x_nuevo,10) for x_nuevo in df['altura_madre']]
sns.lineplot(x=df['altura_madre'],y=prom_vecinos, color="red", alpha=0.5)
prom_vecinos = [predigo_promedio_vecinos(df['altura_madre'],df['altura'],x_nuevo,20) for x_nuevo in df['altura_madre']]
sns.lineplot(x=df['altura_madre'],y=prom_vecinos, color='green')
plt.legend(["k: 4","k: 10","k: 20"])
```
